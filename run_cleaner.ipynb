```python
import pandas as pd
import re
from preprocessing.cleaner import (
    _basic_normalize,
    _mark_sections,
    clean_description,
)

df = pd.read_json("data/raw/sample/sample.jsonl", lines=True)
samples = df["text"].dropna().sample(5, random_state=42)

for i, raw in enumerate(samples):
    t1 = _basic_normalize(raw)
    t2 = _mark_sections(t1)
    
    # 找出所有被识别的章节
    matches = re.findall(r"\|\|SECTION\|\|\s*([^|]+)\|\|", t2)
    sections_found = [m.strip() for m in matches]
    
    final = clean_description(raw)
    
    print(f"\n=== 样本 {i+1} ===")
    print(f"识别到的章节: {sections_found}")
    print(f"最终长度: {len(final)}")
    print(f"前200字: {final[:200]}")
    print("-" * 50)
```

```python
import pandas as pd
from preprocessing.cleaner import _basic_normalize, _extract_kept_sections, clean_description

df = pd.read_json("data/raw/sample/sample.jsonl", lines=True)
samples = df["text"].dropna().sample(5, random_state=42)

for i, raw in enumerate(samples):
    t1 = _basic_normalize(raw)
    extracted = _extract_kept_sections(t1)
    final = clean_description(raw)
    
    print(f"\n=== 样本 {i+1} ===")
    print(f"原始长度:    {len(t1)}")
    print(f"提取后长度:  {len(extracted)}")
    print(f"最终长度:    {len(final)}")
    print(f"提取率:      {len(extracted)/len(t1)*100:.1f}%")
    print(f"最终率:      {len(final)/len(t1)*100:.1f}%")
```

```python
import pandas as pd
from preprocessing.cleaner import _basic_normalize, _extract_kept_sections

df = pd.read_json("data/raw/sample/sample.jsonl", lines=True)
raw = df["text"].dropna().sample(1, random_state=42).iloc[0]

t1 = _basic_normalize(raw)
extracted = _extract_kept_sections(t1)

# 找出被丢弃的内容
original_sentences = set(t1.split(". "))
extracted_sentences = set(extracted.split(". "))

dropped = original_sentences - extracted_sentences

print(f"被丢弃的句子数：{len(dropped)}")
print("\n=== 被丢弃的内容（前10条）===")
for s in list(dropped)[:10]:
    if len(s) > 20:
        print(f"  - {s[:150]}")
```

```python
import pandas as pd
from preprocessing.cleaner import clean_description

# 读取数据
df = pd.read_json("data/raw/sample/sample.jsonl", lines=True)

# 抽取样本
samples = df["text"].dropna().sample(10, random_state=42)

# 清洗并展示
results = []
for i, raw in enumerate(samples):
    cleaned = clean_description(raw)
    results.append({
        "index": i+1,
        "原始长度": len(raw),
        "清洗后长度": len(cleaned),
        "压缩率": f"{len(cleaned)/len(raw)*100:.1f}%",
        "清洗后内容": cleaned
    })

df_results = pd.DataFrame(results)

# 先看统计
print(df_results[["index", "原始长度", "清洗后长度", "压缩率"]])

# 再逐条看内容
for row in results:
    print(f"\n{'='*60}")
    print(f"样本 {row['index']} | 原始:{row['原始长度']} → 清洗后:{row['清洗后长度']} ({row['压缩率']})")
    print(f"{'='*60}")
    print(row["清洗后内容"])
```
